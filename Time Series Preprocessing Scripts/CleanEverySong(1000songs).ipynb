```python
#i = f[1]
f.remove('.DS_Store')

for i in f:

    #print(i)


        
    # Import .csv-file
    df = pd.read_csv (destination+"/"+i,sep=';', header=None)
    

    #artist = df.Artist[0]
    #title = df.Title[0]

    #print(artist)

    #try:
    #    df = df.drop(columns=['Column0', 'Column1','Column2','Column3','Column4','Column5','Column6','Column7'])

    #except:
    #    print("no ekstra columns")

    #df = df.drop(columns=['Artist','ISRC','Duration','Title','City','Country'])

    pivotDf = df

    #df['DateTime'] = df['DateTime'].str[:10]
    #df['Play'] = 1

    df['DateTime'] = df[0].astype('datetime64[ns]')

    pivotDf = pivotDf.pivot_table(index='DateTime')

    pivotDf.fillna(0,inplace=True)
    pivotDf = pivotDf.astype(int)

    #pivotDf.columns = pivotDf.columns.droplevel(0) #remove amount
    pivotDf.columns.name = None               #remove categories
    pivotDf = pivotDf.reset_index()                #index to columns

    pivotDf['DateTime'] = pivotDf['DateTime'].astype('datetime64[ns]')

    minVal = pivotDf['DateTime'].min()

    d = {'Date': [minVal]} #,'TotalPlays':[""]
    emptyDf = pd.DataFrame(data=d)
    emptyDf['Date'] = emptyDf['Date'].astype('datetime64[ns]')

    while len(emptyDf) < 180:
        maxVal = emptyDf['Date'].max()+pd.DateOffset(days=1)
        row = pd.Series([maxVal], index=['Date']) #, 'TotalPlays'
        emptyDf = emptyDf.append(row, ignore_index=True)


    emptyDf.reset_index()
    emptyDf = emptyDf.set_index('Date')
    df = df.set_index('DateTime')

    resultDf = pd.concat([df, emptyDf], axis=1) #, join='inner'
    resultDf.fillna(0,inplace=True)
    resultDf = pd.DataFrame(resultDf[1])
    #resultDf.rename(columns={0: "Artist"})
    
    resultDf = resultDf.astype(int)
    name = i[:-4]
    resultDf.rename(columns={1: str(name)})
    #resultDf = resultDf.drop(columns=0, axis=1)
    resultDf = resultDf.reset_index()
    
    result2Df = resultDf
    result2Df[str(name)] = resultDf[1]
    result2Df = result2Df.drop(columns=['index',1])
    #result2Df.rename(columns={"Artist": str(name)})
    #print(resultDf[1])
    result2Df = result2Df.T
    finalDf = finalDf.append(result2Df, sort=False)
    #finalDf = pd.concat([finalDf, result2Df], axis=1, join='inner') 
    print(result2Df.head())
    
    #resultDf.insert(loc=1, column='ID', value=name)

    #resultDf.to_excel(destination + "/Cleaned/" + name + ".xlsx")

    #resultDf
finalDf = finalDf.iloc[1:]
```
