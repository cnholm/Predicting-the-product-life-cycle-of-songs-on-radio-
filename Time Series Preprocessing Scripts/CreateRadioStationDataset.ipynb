```python
try:
    f.remove('.DS_Store')
except:
    print('nothing')

# Create Empty Dataframe
dfConcat = pd.DataFrame()

for i in f:
    df = pd.read_excel('/Users/clausholm/Documents/Python/Master_thesis/Data/German_radio_cleaned/DailyPlays/All_allRadioStations/' + i, sep=';')
   
    print(i)

    # Set dataset range
    df_range = 28
    df = df.iloc[:df_range]


    df = pd.DataFrame(df.sum())
    df = df.loc['TotalPlays':]

    # Populate new column
    df.insert(0, i[:-5], df[0])
    df = df.drop(columns=[0])

    # Transpose
    df = df.T
    
    # Concatenate Dataframes
    dfConcat = pd.concat([dfConcat, df])
            
    #except:
    #    print("File not found")

dfConcat
```
