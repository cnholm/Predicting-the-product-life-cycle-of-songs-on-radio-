```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/clausholm/Documents/Python/Master_thesis/Data/German_radio_cleaned/DailyPlaysDataset.csv', sep=';', index_col=0)
df
```

```python
df = df.drop(columns=['MaxDailyPlays', 'MaxWeeklyPlays', 'Label'])
df
```

```python
# Create an empty list 
Row_list =[] 
  
# Iterate over each row 
for i in df.iterrows(): 
    
    # append the list to the final list 
    Row_list.append(i) 
  
# Print the list 
print(Row_list) 
```

```python
my_list = df.values.tolist()
my_list
```

```python
all_in_one = []
for i in my_list:
    for ii in i:
        all_in_one.append(ii)

print(all_in_one)
```

```python
import csv
csvfile = "/Users/clausholm/Documents/Python/Master_thesis/Data/German_radio_cleaned/OneLongTimeSeries(DailyPlays).csv"

#Assuming res is a flat list
with open(csvfile, "w") as output:
    writer = csv.writer(output, lineterminator='\n')
    for val in all_in_one:
        writer.writerow([val])   
```
